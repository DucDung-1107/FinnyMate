```python
def create_text_prompt(data, asset_name="Asset", max_entries=10, n=3, open_col='Open', high_col='High', low_col='Low', close_col='Close'):
    prompt = f"Đây là thông tin lịch sử cổ phiếu của {asset_name}. Dự đoán giá cổ phiếu {n} ngày sau:\n\n"
    data = data.tail(max_entries)
    for _, row in data.iterrows():
        # Use the provided column names instead of default ones:
        # Check if the provided column names exist in the DataFrame
        # If not, use default column names
        open_val = row.get(open_col, row.get('Open'))
        high_val = row.get(high_col, row.get('High'))
        low_val = row.get(low_col, row.get('Low'))
        close_val = row.get(close_col, row.get('Close'))

        prompt += f"Date: {row['Date']}, Open: {open_val}, High: {high_val}, Low: {low_val}, Close: {close_val}\n"
    prompt += "\n Gía đóng cửa sẽ là: "
    return prompt

```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

def compute_metrics(actual, predicted):
  #tính r2
  r2 = r2_score(actual, predicted)
  mse= mean_squared_error(actual, predicted)

  rmse = np.sqrt(mse)
  mae = np.mean(np.abs(np.array(actual) - np.array(predicted)))
  return r2, rmse, mae
```

```python
def evaluate_predictions(df, predictions, num_days):
    actual_prices = df['Close'].tail(num_days).tolist()

    # Handle cases where predictions might contain NaN values
    valid_predictions = [p for p in predictions if not np.isnan(p)]

    # Ensure both lists have the same length for comparison
    min_length = min(len(valid_predictions), len(actual_prices))
    valid_actual_prices = actual_prices[:min_length]  # Adjust actual_prices to match
    valid_predictions = valid_predictions[:min_length]  # Adjust predictions to match

    if len(valid_predictions) < 2 : # Not enough valid predictions to compute metrics
        print(f"Not enough valid predictions for {num_days} days. Skipping evaluation.")
        return 0, 0, 0

    r2, rmse, mae = compute_metrics(valid_actual_prices, valid_predictions)
    return r2, rmse, mae
print(evaluate_predictions(df, FPT_predictions, 63))
```

```python
import pandas as pd
import torch
from datasets import Dataset
from transformers import (
    LlamaTokenizer, LlamaForCausalLM,
    DataCollatorForLanguageModeling,
    TrainingArguments, Trainer,
    EvalPrediction
)
import numpy as np
import math

# --- Prompt creation function ---
def create_text_prompt(prompt_data, asset_name="FPT", n=3):
    prompt = f"Asset: {asset_name}\n"

    if not isinstance(prompt_data.index[-1], pd.Timestamp):
        try:
            prompt_data.index = pd.to_datetime(prompt_data.index)
        except:
            prompt += f"Predict the closing price {n} days after the last available date.\n"

    if isinstance(prompt_data.index[-1], pd.Timestamp):
        prompt += f"Predict the closing price {n} days after {prompt_data.index[-1].strftime('%Y-%m-%d')}.\n"

    prompt += "Recent Data:\n"
    for date, row in prompt_data.iterrows():
        prompt += f"  {date.strftime('%Y-%m-%d')}: Close={row['Close']:.2f}, Vol={row['Volume']}\n"
    prompt += "Prediction:"
    return prompt

# --- Create dataset from DF ---
def create_dataset_FPT_from_df(df, num_days_future=3):
    dataset = []
    max_index = len(df) - num_days_future - 1
    for i in range(len(df) - 10 - num_days_future + 1):
        if i + 10 + num_days_future > len(df):
            break
        prompt_data = df.iloc[i: i + 10]
        label_index = i + 10 + num_days_future - 1
        if label_index >= len(df):
            continue

        prompt_text = create_text_prompt(prompt_data, asset_name="FPT", n=num_days_future)
        label_value = df['Close'].iloc[label_index]
        label_text = f"{label_value:.4f}"
        full_text = prompt_text + " " + label_text + tokenizer.eos_token

        dataset.append({'text': full_text, 'prompt': prompt_text, 'label': label_text})
    return dataset

# --- Load Tokenizer & Model ---

if tokenizer.pad_token is None:
    tokenizer.add_special_tokens({'pad_token': '[PAD]'})
if tokenizer.pad_token_id is not None:
    model.config.pad_token_id = tokenizer.pad_token_id
else:
    model.config.pad_token_id = tokenizer.eos_token_id


# --- Dataset prep ---
num_days_future = 3
FPT_dataset_list = create_dataset_FPT_from_df(df, num_days_future=num_days_future)
hf_dataset = Dataset.from_list(FPT_dataset_list)
hf_dataset = hf_dataset.train_test_split(test_size=0.2)
train_dataset = hf_dataset["train"]
eval_dataset = hf_dataset["test"]

# --- Tokenization ---
def tokenize_function(examples):
    return tokenizer(examples["text"], truncation=True)

tokenized_train = train_dataset.map(tokenize_function, batched=True, remove_columns=['text', 'prompt', 'label'])
tokenized_eval = eval_dataset.map(tokenize_function, batched=True, remove_columns=['text', 'prompt', 'label'])

# --- Metric: Perplexity ---
def compute_metrics(eval_pred: EvalPrediction):
    loss = eval_pred.loss
    return {"perplexity": math.exp(loss)}

# --- Data collator ---
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)

# --- Training Arguments ---
training_args = TrainingArguments(
    output_dir="./llama-fpt-finetuned",
    num_train_epochs=10,
    per_device_train_batch_size=1,
    learning_rate=2e-5,
    logging_dir='./logs',
    logging_steps=10,
    save_total_limit=2,
)


# --- Trainer ---
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_eval,
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)

# --- Train and Evaluate ---
print("Starting fine-tuning...")
trainer.train()
print("Fine-tuning complete.")

# --- Save model and tokenizer ---
final_model_path = "./llama-fpt-finetuned-final"
trainer.save_model(final_model_path)
tokenizer.save_pretrained(final_model_path)
print(f"Model saved to {final_model_path}")

# --- Example Inference ---
def generate_prediction(prompt_text, max_new_tokens=10):
    inputs = tokenizer(prompt_text, return_tensors="pt").to(model.device)
    outputs = model.generate(**inputs, max_new_tokens=max_new_tokens)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Use last 10 rows for example prediction
example_prompt = create_text_prompt(df.iloc[-10:], asset_name="FPT", n=3)
print("\n--- Example Prompt ---")
print(example_prompt)
print("\n--- Model Prediction ---")
print(generate_prediction(example_prompt))
```
