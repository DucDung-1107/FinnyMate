```python

def find_outliers_iqr(data, column_name):
    Q1 = data[column_name].quantile(0.2)
    Q3 = data[column_name].quantile(0.8)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = data[(data[column_name] < lower_bound) | (data[column_name] > upper_bound)]
    num_outliers = len(outliers)

    return num_outliers, outliers

# Example usage for FPT close prices:
num_outliers_fpt, outlier_fpt = find_outliers_iqr(fpt_data, 'Close_FPT')
#vẽ boxplot
sns.boxplot(x=fpt_data['Close_FPT'])
plt.show()
print(f"Number of outliers in FPT Close Price: {num_outliers_fpt}")
print("Outliers:")
print(outlier_fpt)



```

```python
num_outliers_cmc, outlier_cmc = find_outliers_iqr(cmc_data, 'Close_CMC')
#vẽ boxplot
sns.boxplot(x=cmc_data['Close_CMC'])
plt.show()
print(f"Number of outliers in cmc Close Price: {num_outliers_cmc}")
print("Outliers:")
print(outlier_cmc)
```

```python

def count_outliers_pct_change(data, column_name, threshold=0.15):

    # Calculate percentage change
    data['pct_change'] = data[column_name].pct_change()

    # Count outliers where the absolute percentage change exceeds the threshold
    outliers = data[abs(data['pct_change']) > threshold]
    num_outliers = len(outliers)

    return num_outliers

# Example usage for FPT Close Prices:
num_outliers_fpt = count_outliers_pct_change(fpt_data, 'Close_FPT')
print(f"Number of outliers in FPT Close Price (pct_change > 15%): {num_outliers_fpt}")

# Example usage for CMC Close Prices:
num_outliers_cmc = count_outliers_pct_change(cmc_data, 'Close_CMC')
print(f"Number of outliers in CMC Close Price (pct_change > 15%): {num_outliers_cmc}")
```
